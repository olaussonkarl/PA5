```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np

data_gen = ImageDataGenerator(rescale=1.0/255)
```

```python
imgdir = 'a5_images'
img_size = 64
batch_size = 32

train_generator = data_gen.flow_from_directory(
        imgdir + '/train',
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='binary',
        classes=['other', 'car'],
        seed=12345,
        shuffle=True)
```

```python
val_generator = data_gen.flow_from_directory(
        imgdir + '/validation',
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='binary',
        classes=['other', 'car'],
        seed=12345,
        shuffle=True)
```

```python
import tensorflow as tf 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
```

```python
def make_convnet(img_size):
    model = Sequential()
    
    model.add(Conv2D(32, (3,3), 1, activation='relu', input_shape=(img_size, img_size, 3)))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(64, (3,3), 1, activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Conv2D(32, (3,3), 1, activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    
    model.add(Flatten())
    
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    model.compile(loss=tf.losses.BinaryCrossentropy(),
                 optimizer="adam",
                 metrics=['accuracy'])
    
    return model
```

```python
model1 = make_convnet(img_size)
```

```python
hist = model1.fit(train_generator, batch_size=batch_size,epochs=20, validation_data=val_generator)
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='blue', label='loss')
plt.plot(hist.history['val_loss'], color='yellow', label='val_loss')
fig.suptitle('loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='blue', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='yellow', label='val_accuracy')
fig.suptitle('loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
from tensorflow.keras import applications
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.applications.vgg16 import decode_predictions, preprocess_input

vggmodel = applications.VGG16(weights='imagenet', include_top=True)
```

```python
feature_extractor = applications.VGG16(include_top=False, weights='imagenet', input_shape=(img_size, img_size, 3))
```

```python
vgg_data_gen = ImageDataGenerator(preprocessing_function=preprocess_input)
```

```python
def create_vgg16_features(train_or_test, some_file_name):
   
    train_generator2 = vgg_data_gen.flow_from_directory(
        imgdir + train_or_test,
        target_size=(img_size, img_size),
        batch_size=batch_size,
        class_mode='binary',
        classes=['other', 'car'],
        seed=12345,
        shuffle=False)
    
    cnn_features = feature_extractor.predict(train_generator2)
    
    with open(some_file_name, 'wb') as f:
        np.save(f, cnn_features)
    
```

```python
create_vgg16_features('/train', "Training.txt")
```

```python
create_vgg16_features('/validation', "Validations.txt")
```

```python
def get_labels(n):
    return np.array([0]*(n//2) + [1]*(n//2))
```

```python
def train_on_cnnfeatures(training_file, validation_file):
    
    with open(training_file, 'rb') as f:
        train_ds = np.load(f)
    
    with open(validation_file, 'rb') as f:
        validation_ds = np.load(f)
    
    labels_train = get_labels(len(train_ds))
    
    labels_val = get_labels(len(validation_ds))
    
    model = Sequential()
    
    model.add(Flatten())
    
    model.add(Dense(64, activation='relu'))
    model.add(Dense(1, activation='softmax'))
    
    model.compile(
        loss='categorical_crossentropy',
        optimizer='adam',
        metrics=['accuracy']
    )
    
    hist = model.fit(train_ds, labels_train, batch_size=batch_size, epochs=20, validation_data=(validation_ds, labels_val))
    
    return hist
```

```python
hist = train_on_cnnfeatures('Training.txt', 'Validations.txt')
```
